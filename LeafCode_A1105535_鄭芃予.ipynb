```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
xy = np.array([[0],[0]])
```

```python
t1 = [np.array([[0.86,0.03],[-0.03,0.86]]), np.array([[0],[1.5]])]
```

```python
t2 = [np.array([[0.2,-0.25],[0.21,0.23]]),np.array([[0],[1.5]])]
```

```python
t3 = [np.array([[-0.15,0.27],[0.25,0.26]]),np.array([[0],[0.45]])]
```

```python
t4 = [np.array([[0,0],[0,0.17]]),np.array([[0],[0]])]
```

```python
t_list = [t1,t2,t3,t4] 
```

```python
p1 = 0.83
p2 = 0.08
p3 = 0.08
p4 = 0.01
p_list = (p1,p2,p3,p4)
```

```python
plt.plot(xy[0],xy[1])
for i in range(5000):
    t = random.choices(t_list, weights=p_list,k=1)
    xy = np.matmul(t[0][0],xy)+t[0][1]
    plt.scatter(xy[0],xy[1],color = 'g',s=5)
```
